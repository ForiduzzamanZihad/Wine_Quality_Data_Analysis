```python
x=data.iloc[:,:].values
y=data.iloc[:,-1].values
y
```

Balancing the imbalancing data of Target feature

```python
from imblearn.over_sampling import SMOTE
S=SMOTE()
x_data,y_data=S.fit_resample(x,y)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
S=MinMaxScaler()
X=S.fit_transform(x_data)
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y_data,random_state=250,test_size=0.25)
```

Here we use KNN (KNeighborsClassifier)

```python
from sklearn.neighbors import KNeighborsClassifier
II=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=1)
II.fit(x_train,y_train)
```

```python
y_pred=II.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)*100
print(ac)
```
